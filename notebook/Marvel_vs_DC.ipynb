```python
import pandas as  pd
import requests
```

## WEBSCRAPPING MARVEL DATABASE

```python
url_marvel = "https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}


response = requests.get(url_marvel, headers=headers)

tables = pd.read_html(response.text)

mcu_movies = tables[0]

print(f"Total tables found: {len(tables)}")

# Loop through tables to see which one has the columns you want
for i, table in enumerate(tables):
    print(f"\nTable {i} columns:\n", table.columns)
```

## Extracting needed tables

```python
mcu_table = tables[10]  # Table with Box office and Budget
ratings_table = tables[11]  # Table with critic/public ratings

print(ratings_table)
```

```python
# Flatten MultiIndex columns MCU TABLE
mcu_table.columns = [' '.join(col).strip() if type(col) is tuple else col for col in mcu_table.columns]
print(mcu_table.head())
```

```python
ratings_table.columns = [' '.join(col).strip() if type(col) is tuple else col for col in ratings_table.columns] # Ratings tables
print(ratings_table.head())
```

CHNAGING NAMES OF THE COLUMNS

```python
mcu_table = mcu_table.rename(columns={
    'Film Film Phase One': 'Movie',
    'U.S. release date U.S. release date Phase One': 'Release_Date',
    'Box office gross Worldwide Phase One': 'Box_Office',
    'Budget Budget Phase One': 'Budget',
    'All-time ranking Worldwide Phase One': 'World_Rankings',
    'Budget Budget Phase On': 'Budget'
})

print(mcu_table.head())
```

With Ratings table


```python
ratings_table1 = ratings_table.rename(columns={
    'Film Film Phase One': 'Movie',
    'Critical Rotten Tomatoes Phase One': 'Rotten_Tomatoes',
})

Ratings_clean = ratings_table1[['Movie', 'Rotten_Tomatoes']]

print(Ratings_clean.head())
```

## SELECTING RELEVANT COLUMNS

```python
mcu_clean = mcu_table[['Movie', 'Release_Date', 'Budget', 'Box_Office']]
mcu_clean['Franchise'] = 'Marvel'
print(mcu_clean.head())
```

Merging two tables of MCU franchise



```python
M_Clean = M_Clean = pd.merge(
    mcu_clean,         # main table
    Ratings_clean,     # ratings table
    on="Movie",         # common column
    how="left"         # keep all films, add ratings where available
)

M_Clean = M_Clean.drop(M_Clean.index[-1]) # dropping last row as it is total
print(M_Clean.head(50))
```

```python
M_Clean = M_Clean[~M_Clean['Movie'].str.contains('Phase', na=False)]



M_Clean = M_Clean.reset_index(drop=True)

M_Clean.index = M_Clean.index + 1



M_Clean
```

# WEB SCRAPING FOR DCEU MOVIES

```python
url_DCEU = "https://en.wikipedia.org/wiki/DC_Extended_Universe"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}


response = requests.get(url_DCEU, headers=headers)

tables1 = pd.read_html(response.text)


print(f"Total tables found: {len(tables1)}")

# Loop through tables to see which one has the columns you want
for i, table2 in enumerate(tables1):
    print(f"\nTable {i} columns:\n", table2.columns)
```

```python
dcu_table = tables1[4]  # Table with Box office and Budget
dratings_table = tables1[5]  # Table with critic/public ratings

dcu_table.columns = [' '.join(col).strip() if type(col) is tuple else col for col in dcu_table.columns]


dratings_table.columns = [' '.join(col).strip() if type(col) is tuple else col for col in dratings_table.columns]

dcu_table = dcu_table.rename(columns={
    'Film Film': 'Movie',
    'U.S. release date U.S. release date': 'Release_Date',
    'Box office gross Worldwide': 'Box_Office',
    'Budget Budget': 'Budget'
})

dratings_table = dratings_table.rename(columns={'Title Title' : 'Movie', 'Critical Rotten Tomatoes' : 'Rotten_Tomatoes'})

dcu_table = dcu_table[['Movie', 'Release_Date', 'Box_Office', 'Budget']]
dcu_table['Franchise'] = 'DC'

dratings_table = dratings_table[['Movie', 'Rotten_Tomatoes']]

D_Clean = pd.merge(
    dcu_table,
    dratings_table,
    on='Movie',
    how='left')

D_Clean = D_Clean.drop(D_Clean.index[-1])
print(D_Clean.tail())
print(D_Clean.shape)
```

```python
url_DCU = "https://en.wikipedia.org/wiki/DC_Universe_(franchise)"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}


response = requests.get(url_DCU, headers=headers)

tables2 = pd.read_html(response.text)


print(f"Total tables found: {len(tables2)}")

# Loop through tables to see which one has the columns you want
for i, table3 in enumerate(tables2):
    print(f"\nTable {i} columns:\n", table3.columns)
```

```python
dc_table = tables2[4]  # Table with Box office and Budget
drating_table = tables2[5]  # Table with critic/public ratings

dc_table.columns = [' '.join(col).strip() if type(col) is tuple else col for col in dc_table.columns]


drating_table.columns = [' '.join(col).strip() if type(col) is tuple else col for col in drating_table.columns]

dc_table = dc_table.rename(columns={
    'Film Film': 'Movie',
    'U.S. release date U.S. release date': 'Release_Date',
    'Box office gross Worldwide': 'Box_Office',
    'Budget Budget': 'Budget'
})

drating_table = drating_table.rename(columns={'Title Title' : 'Movie', 'Critical Rotten Tomatoes' : 'Rotten_Tomatoes'})


dc_table = dc_table[['Movie', 'Release_Date', 'Box_Office', 'Budget']]
dc_table['Franchise'] = 'DC'

drating_table = drating_table[['Movie', 'Rotten_Tomatoes']]

dc_table.index = dc_table.index + 1

dc_table = pd.merge(dc_table,
                    drating_table,
                    on='Movie',
                    how='left')
dc_table = dc_table.drop(dc_table.index[-1])
dc_table.index = dc_table.index + 1

print(dc_table)
dc_table.shape
```

```python
import re

def parse_budget(value):
    if pd.isna(value):
        return None
    # Remove $, commas, the word "million", and NBSP
    value = re.sub(r"[^\d\.\–-]", "", value)

    # If it's a range, take the average
    if "–" in value or "-" in value:
        sep = "–" if "–" in value else "-"
        parts = value.split(sep)
        nums = [float(p) for p in parts if p]
        return sum(nums) / len(nums) * 1_000_000
    else:
        return float(value) * 1_000_000

Movies_clean["Budget"] = Movies_clean["Budget"].apply(parse_budget)

```

```python

def parse(x):
    if pd.isna(x):
        return None
    # Remove $, commas, the word "million", and NBSP
    x = re.sub(r'\[.*?\]', '', str(x))  # remove [ ... ]
    x = x.replace('$', '').replace(',', '').strip()
    return int(x) if x.isdigit() else int(float(x))

Movies_clean["Box_Office"] = Movies_clean["Box_Office"].apply(parse)
```
